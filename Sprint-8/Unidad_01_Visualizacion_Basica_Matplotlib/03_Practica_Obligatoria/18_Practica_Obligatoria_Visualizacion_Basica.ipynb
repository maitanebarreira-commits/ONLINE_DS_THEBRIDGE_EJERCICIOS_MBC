#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python

# Crear figura 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Grfico 1: Histograma
sns.histplot(df["elevation_m"], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribucin de la altitud (m)")
axes[0, 0].set_xlabel("Altitud (m)")
axes[0, 0].set_ylabel("Frecuencia")

# Grfico 2: Scatter
axes[0, 1].scatter(df["elevation_m"], df["population_total"], alpha=0.6, edgecolor="k")
axes[0, 1].set_title("Relacin entre altitud y poblacin")
axes[0, 1].set_xlabel("Altitud (m)")
axes[0, 1].set_ylabel("Poblacin total")

# Grfico 3: Boxplot
sns.boxplot(x=df["area_total_km2"], ax=axes[1, 0])
axes[1, 0].set_title("Boxplot del rea total (km)")
axes[1, 0].set_xlabel("rea total (km)")

# Grfico 4: Latitud vs Longitud
axes[1, 1].scatter(df["longd"], df["latd"], alpha=0.6)
axes[1, 1].set_title("Ubicacin de las ciudades")
axes[1, 1].set_xlabel("Longitud")
axes[1, 1].set_ylabel("Latitud")

```
